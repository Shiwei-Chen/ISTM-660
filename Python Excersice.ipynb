```
# 1
plaintext = input('enter a word: ')
distance = int(input('enter distance: '))

code = ''
for ch in plaintext:
  ordvalue = ord(ch)
  ciphervalue = ordvalue + distance

  if ciphervalue > ord('z'):
    ciphervalue = ord('a') + distance - (ord('z') - ordvalue + 1)
  elif ord('Z') < ciphervalue < ord('a'):
    ciphervalue = ord('A') + distance - (ord('Z') - ordvalue + 1)
  code += chr(ciphervalue)
  
print(code)
```

```
#9 There is a numerical method to calculate square root, it is called Newton's method. Look into the algorithm and try to implement it using regular and recursive method.
import math
tolerance = 0.000001
```

```
#9 continued
def newton(x):
  '''returns the square root of x.'''
  #perform approximation
  estimate = 1.0
  a = True
  while a:
    estimate = (estimate + x/estimate)/2
    difference = abs(x-estimate**2)
    if difference <= tolerance:
      a = False
  return estimate
```

```
from breezypythongui import EasyFrame

def computeDistance(height, index, tolerance):
    """Computes the total distance traveled by the ball,
    given an initial height, bounciness index, and
    number of tolerance."""
    total = 0
    n = 1
    while height >= tolerance:
        total += height
        n += 1
        height *= index
    return total

def computeBounces(height, index, tolerance):
    total = 0
    n = 1
    while height >= tolerance:
        total += height
        n += 1
        height *= index
    return n


class BouncyGUI(EasyFrame):

    def __init__(self):
        """Set up the window and widgets."""
        EasyFrame.__init__(self,  title = "Bouncy")
        self.addLabel(text = "Initial height", row = 0, column = 0)
        self.heightField = self.addFloatField(value = 0.0, row = 0, column = 1)
        self.addLabel(text = "Bounciness index", row = 1, column = 0)
        self.indexField = self.addFloatField(value = 0.0, row = 1, column = 1)
        self.addLabel(text = "Tolerance", row = 2, column = 0)
        self.toleranceField = self.addFloatField(value = 0.0000, row = 2, column = 1)
        self.addButton(text = "Compute", row = 3, column = 0,
                       columnspan = 2, command = self.computeDistance)
        self.addLabel(text = "Total distance", row = 4, column = 0)
        self.distanceField = self.addFloatField(value = 0.0, row = 4, column = 1)
        self.addLabel(text = "Bounces", row = 5, column = 0)
        self.bouncesField = self.addFloatField(value = 0, row = 5, column = 1)

    def computeDistance(self):
        """Event handler for the Compute button."""
        height = self.heightField.getNumber()
        index = self.indexField.getNumber()
        tolerance = self.toleranceField.getNumber()
        distance = computeDistance(height, index, tolerance)
        self.distanceField.setNumber(distance)
        bounce = computeBounces(height, index, tolerance)
        self.bouncesField.setNumber(bounce)


def main():
    """Instantiate and pop up the window."""
    BouncyGUI().mainloop()

if __name__ == "__main__":
    main()
```
